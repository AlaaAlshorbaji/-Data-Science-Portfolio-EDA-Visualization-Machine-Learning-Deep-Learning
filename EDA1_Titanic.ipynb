# **EDA**

## **UNIVARIATE ANALYSIS**

```python
#histogram data ....continuous
plt.hist(df['Age'])
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

sns.histplot(df['Age'])
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Density curve (KDE):,numerical data ....continuous
sns.kdeplot(df['Age'], shade=True,color='blue')
plt.title('Density Curve')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
#boxplot data ....continuous
sns.boxplot(x=df['Age'])
plt.title('Boxplot')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#violin: data ....continuous
sns.violinplot(x=df['Age'])
plt.title('Violin Plot')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar plot :categorical data
sns.countplot(x=df['Sex'])
plt.title('Bar Plot')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
#pie chart :
labels = ['Survived', 'Not Survived']
#sizes =
colors = ['green', 'red']
explode = (0.1, 0)
plt.pie(df['Survived'].value_counts(),labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

## **Bivariate Analysis**

```python
#scatter plots : num vs num :
plt.scatter(df['Age'], df['Fare'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.scatterplot(x=df['Age'], y=df['Fare'],hue=df['Survived'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#line plot : num vs num
plt.plot(df['Age'], df['Fare'])
plt.title('Line Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.lineplot(x=df['Age'], y=df['Fare'],hue=df['Survived'])
plt.title('Line Plot')
plt.xlabel('Age')

plt.ylabel('Fare')
plt.show()
```

```python
#regression plot :
sns.regplot(x=df['Age'], y=df['Fare'])
plt.title('Regression Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#pair plot :
sns.pairplot(df)
plt.show()
```

```python
#bar plot : cat vs cat/discrete:
sns.barplot(x=df['Sex'], y=df['Survived'])
plt.title('Bar Plot')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar plot : cat vs cat :
sns.barplot(x=df['Sex'], y=df['Pclass'], hue=df['Survived'])
plt.title('Bar Plot')
plt.xlabel('Sex')
plt.ylabel('Pclass')
plt.show()
```

```python
#boxplot cat vs num :
sns.boxplot(x=df['Sex'], y=df['Age'])
plt.title('Box Plot')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
#boxplot cat vs num :
sns.boxplot(x=df['Sex'], y=df['Fare'],hue=df['Survived'])
plt.title('Box Plot')
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.show()
```

```python
#violin plot :
sns.violinplot(x=df['Sex'], y=df['Age'],hue=df['Survived'])
plt.title('Violin Plot')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
#swarm :
sns.swarmplot(x=df['Sex'], y=df['Age'],hue=df['Survived'])
plt.title('Swarm Plot')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
